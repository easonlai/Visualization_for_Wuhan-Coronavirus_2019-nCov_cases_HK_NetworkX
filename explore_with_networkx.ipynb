# Relationship visualization of Wuhan Coronavirus (2019-nCov) cases in Hong Kong

Download open data "Residential buildings in which probable/confirmed cases have resided in the past 14 days or non-residential building with 2 or more probable/confirmed cases in the past 14 days (English)" & "Details of probable/confirmed cases of COVID-19 infection in Hong Kong (English)" from Centre for Health Protection Hong Kong (CHP) to build relationship dataset.

Other open data from CHP can be obtain from https://data.gov.hk/en-data/dataset/hk-dh-chpsebcddr-novel-infectious-agent

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
hkcase_data_2 = pd.read_csv('hkcase_relationship.csv')
hkcase_data_2 = hkcase_data_2.dropna()
hkcase_data_2 = hkcase_data_2.astype(int)
```

```python
G = nx.Graph() 

plt.figure(figsize =(25, 25))

G.add_edges_from(hkcase_data_2.values) 

# Show directed graph
H = nx.to_directed(G) 

pos = nx.spring_layout(G, k=1*1/np.sqrt(len(G.nodes())), iterations=20)
plt.figure(3, figsize=(20, 20))
nx.draw(G, pos=pos)
# Show directed graph
#nx.draw(H, pos=pos)
nx.draw_networkx_labels(G, pos=pos)
plt.show()
```
